```python
Y = df['label'].to_numpy(); 
df.drop(columns = ['label'],inplace=True)
```

```python
X = df.to_numpy()
```

```python
def Pca(X,k): 
    mean = np.mean(X,axis=0); 
    Xmean = X-mean; 
    N = X.shape[0]; 
    cov = 1/N * (Xmean.T @ Xmean); 
    val,vec = np.linalg.eigh(cov); 
    index = np.argsort(val)[::-1][:k] 
    pca = vec[:,index];
    s = np.sum(val[index])/np.sum(val);
    return Xmean@pca,s
```

```python
XX,_ = Pca(X,50);
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from lightgbm import LGBMClassifier
```

```python
x_train,x_test,y_train,y_test = train_test_split(XX,Y,test_size=0.1,shuffle=True); 
```

```python
model1 = LGBMClassifier(verbose=0,n_estimators=2000); 
model1.fit(x_train,y_train); 
predict1 = model1.predict(x_test); 
print(accuracy_score(y_test,predict1))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model2 = RandomForestClassifier(n_estimators=1000);
model2.fit(x_train,y_train); 
predict2 = model2.predict(x_test);
print(accuracy_score(y_test,predict2))
```

```python
from catboost import CatBoostClassifier
model3 = CatBoostClassifier(verbose=0,n_estimators=1000); 
model3.fit(x_train,y_train); 
predict3 = model3.predict(x_test);
print(accuracy_score(y_test,predict3));
```

```python
from sklearn.model_selection import KFold 
kf = KFold(n_splits=5);
models = [] 
acc = [] 
for train_index,test_index in kf.split(X): 
    model = LGBMClassifier(verbose=-1,n_estimators=1000); 
    model.fit(X[train_index],Y[train_index]); 
    predict = model.predict(X[test_index]); 
    acc.append(accuracy_score(Y[test_index],predict)); 
    models.append(model);
```
